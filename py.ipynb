```python
from PIL import Image

def encode_text(image_path, text, lsb):
    # Open the image
    image = Image.open(image_path)
    pixels = image.load()

    # Convert text to binary
    binary_text = ''.join(format(ord(char), '08b') for char in text)

    # Check if the image can hold the text
    max_chars = (image.size[0] * image.size[1] * lsb) // 8
    if len(binary_text) > max_chars:
        raise ValueError("Text is too long to be encoded in the image.")

    # Encode the text in the image
    char_index = 0
    for y in range(image.size[1]):
        for x in range(image.size[0]):
            r, g, b = pixels[x, y]

            # Modify the least significant bit of each color channel
            if char_index < len(binary_text):
                r = (r & ~(1 << lsb)) | (int(binary_text[char_index]) << lsb)
                char_index += 1
            if char_index < len(binary_text):
                g = (g & ~(1 << lsb)) | (int(binary_text[char_index]) << lsb)
                char_index += 1
            if char_index < len(binary_text):
                b = (b & ~(1 << lsb)) | (int(binary_text[char_index]) << lsb)
                char_index += 1

            pixels[x, y] = (r, g, b)

    # Save the modified image
    encoded_image_path = "encoded_image7.png"
    image.save(encoded_image_path)
    print("Text encoded successfully in the image:", encoded_image_path)

def decode_text(image_path, lsb):
    # Open the image
    image = Image.open(image_path)
    pixels = image.load()

    # Decode the text from the image
    binary_text = ""
    for y in range(image.size[1]):
        for x in range(image.size[0]):
            r, g, b = pixels[x, y]

            # Extract the least significant bit of each color channel
            binary_text += str((r >> lsb) & 1)
            binary_text += str((g >> lsb) & 1)
            binary_text += str((b >> lsb) & 1)

    # Convert binary text to ASCII
    decoded_text = ""
    for i in range(0, len(binary_text), 8):
        decoded_text += chr(int(binary_text[i:i+8], 2))

    return decoded_text

# Example usage
image_path = input("Enter the image location: ")
lsb = int(input("Enter the number of LSBs to use (1-8): "))
text = input("Enter the text to encode: ")

encode_text(image_path, text, lsb)

decoded_text = decode_text("encoded_image7.png", lsb)
print("Decoded text:", decoded_text)
```
